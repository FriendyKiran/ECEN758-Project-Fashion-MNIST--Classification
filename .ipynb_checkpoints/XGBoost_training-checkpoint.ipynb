```python
from sklearn.datasets import fetch_openml
import pandas as pd
```

```python
#Load Fashion-Mnist Dataset
fashion_mnist = fetch_openml(name="Fashion-MNIST", cache = True)
fashion_mnist_df = pd.DataFrame(fashion_mnist.data, columns=fashion_mnist.feature_names)
fashion_mnist_df['target'] = fashion_mnist['target'].astype(int)
fashion_mnist_df.head(10)
```

```python
fashion_mnist_df.shape
```

```python
X = fashion_mnist_df.drop('target', axis = 1).values #Get the data 28x28 total 784 features
y = fashion_mnist_df['target'].values #Get the target values or class
```

```python
#Splitting the data into training and testing set
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
#Coverting the scale of pixels from 0-1
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
# PCA

from sklearn.decomposition import PCA

pca = PCA(n_components = 0.95)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print("X_train shape after Dimensionality Reduction: ",X_train_pca.shape)
print("X_test shape after Dimensionality Reduction: ", X_test_pca.shape)
```

```python
# XGBoost Classifier

import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score

parameters = [{
    'max_depth': [3, 4, 5], # Maximum depth of a tree.
    'learning_rate': [0.001, 0.003, 0.01, 0.03, 0.1, 0.2], # Step size shrinkage used to prevent overfitting.
    'gamma': [0, 0.25, 0.5], # min_split_loss : Minimum loss reduction required to make a further partition on a leaf node of the tree.
    'subsample': [0.6, 0.7, 0.8] # Subsample ratio of the training instances for each tree
}]

xgb_clf = xgb.XGBClassifier(tree_method='gpu_hist')

random_search_xgb = RandomizedSearchCV(
    estimator = xgb_clf,
    param_distributions = parameters,
    n_iter=50,
    cv = 5,
    random_state=5,
    scoring=['accuracy','f1','recall','precision'],
    verbose=10,
    n_jobs = -1,
    refit = 'accuracy')

random_search_xgb.fit(X_train_pca, y_train)


best_params_xgb = random_search_xgb.best_params_
best_accuracy_xgb = random_search_xgb.best_score_
best_xgb = random_search_xgb.best_estimator_

# Evaluate the model on the test set
y_pred = best_xgb.predict(X_test_pca)
accuracy_xgb = accuracy_score(y_test, y_pred)


print("---------------------------------------")
print("Best Parameters:", best_params_xgb)
print("---------------------------------------")
print("Best Accuracy in Cross Validation:", best_accuracy_xgb)
print("Test Accuracy:", accuracy_xgb)
```

```python
cv_results = random_search_xgb.cv_results_
print(cv_results.keys())
results_xgb = {}
for accuracy, params in zip(cv_results["mean_test_accuracy"],cv_results["params"]):
    results_xgb.update({str(params) : accuracy})
```

```python
import pandas as pd
df_xgb = pd.DataFrame.from_dict(results_xgb, orient = 'index')
```

```python
df_xgb.reset_index(inplace=True)
```

```python
df_xgb.columns = ['Parameters','Accuracy']
```

```python
df_xgb.sort_values('Accuracy', ascending = False)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split

X_train_xgb, X_val, y_train_xgb, y_val = train_test_split(X_train_pca, y_train, test_size=0.2)

# Create DMatrices - XGBoost's internal data structure, optimized for both memory efficiency and training speed
dtrain = xgb.DMatrix(X_train_xgb, label=y_train_xgb)
dval = xgb.DMatrix(X_val, label=y_val)

# Using the best parameters for the XGB model from RandomSearchCV
params = {'subsample': 0.7, 'max_depth': 5, 'learning_rate': 0.2, 'gamma': 0.5,
          'tree_method': 'gpu_hist', 'objective': 'multi:softmax', 'num_class': 10}

evals_result = {}

# Train the model
bst = xgb.train(params, dtrain, num_boost_round=100, evals=[(dtrain, 'train'), (dval, 'eval')], evals_result=evals_result)
```

```python
train_error_xgb = evals_result['train']['mlogloss']
val_error_xgb = evals_result['eval']['mlogloss']
```

```python
import matplotlib.pyplot as plt

iterations = range(1, len(train_error_xgb) + 1)

plt.figure(figsize=(8, 6))
plt.plot(iterations, train_error_xgb, label='Train')
plt.plot(iterations, val_error_xgb, label='Validation')

plt.title('Training and Validation Loss - XGBoost Classifier')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import pickle

# Save the model to a file
with open('xgboost_model.pkl', 'wb') as file:
    pickle.dump(bst, file)
```
